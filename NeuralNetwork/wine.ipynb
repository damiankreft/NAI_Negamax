```python
import tensorflow as tf
from keras import models
from keras.models import Sequential
from keras.layers import Dense,Dropout, BatchNormalization
from keras import regularizers, layers
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import tensorflow_addons as tfa

original_df = pd.read_csv('datasets/WineQT.csv')
original_df = original_df.drop('Id',axis =1)
original_df.columns

# raw_val_ds = tf.keras.utils.text_dataset_from_directory(
#     'aclImdb/train', 
#     validation_split=0.2, 
#     subset='validation', 
#     seed=42)
```

```python
original_df.head(10)
```

```python
wine_features = original_df.copy()
wine_labels = wine_features.pop('quality')
wine_features
```

```python
wine_model = tf.keras.Sequential([
  layers.Activation(activation='relu'),
  layers.Dense(64),
  layers.Dense(1),
])


wine_model.compile(loss = tf.keras.losses.MeanSquaredError(),
                      optimizer = tf.keras.optimizers.Adam())
```

```python
wine_model.fit(wine_features, wine_labels, epochs=10)
```

```python
normalize = layers.Normalization()
```

```python
normalize.adapt(wine_features)
```

```python
norm_wine_model = tf.keras.Sequential([
  normalize,
  layers.Dense(32),
  layers.Dense(1)
])

norm_wine_model.compile(loss = tf.keras.losses.MeanSquaredError(),
                        metrics=tf.metrics.BinaryAccuracy(threshold=0.0),
                           optimizer = tf.keras.optimizers.Adam())

norm_wine_model.fit(wine_features, wine_labels, epochs=10)
```
